```python
files_PLS = ['PLS-iron.csv', 'PLS-rebar.csv']
files_PCR = ['PCR-iron.csv', 'PCR-rebar.csv']
```

```python
import statsmodels.api as sm

for file1, file2 in zip(files_PLS, files_PCR):
    print(file2.split('-')[1][:-4])
    df_metric = pd.read_csv('/home/yicheng.wang/UCL-Dissertation/Outputs/ML-model/PLS/' + file1)

    # df_metric = df_metric[df_metric['lookback'] <= 96]
    df = df_metric[['lookback','horizon','mse']]

    X = df.lookback
    y = df.mse

    X = sm.add_constant(X)

    est=sm.OLS(y, X)
    est = est.fit()
    print(est.summary())


    X_prime = np.linspace(X.lookback.min(), X.lookback.max(), 100)[:, np.newaxis] 
    X_prime = sm.add_constant(X_prime) 

    y_hat = est.predict(X_prime) 

    
    df_metric2 = pd.read_csv('/home/yicheng.wang/UCL-Dissertation/Outputs/ML-model/PCR/' + file2)
    # df_metric = df_metric[df_metric['lookback'].isin([12,24,36,48,60,72,84,96,108,120])]
    
    # df_metric2 = df_metric2[df_metric2['lookback'] <= 96]
    df2 = df_metric2[['lookback','horizon','mse']]
    X2 = df2.lookback
    y2 = df2.mse
    X2 = sm.add_constant(X2)

    est2=sm.OLS(y2, X2)
    est2 = est2.fit()
    print(est2.summary())


    X_prime2 = np.linspace(X2.lookback.min(), X2.lookback.max(), 100)[:, np.newaxis] 
    X_prime2 = sm.add_constant(X_prime2) 
    y_hat2 = est2.predict(X_prime2) 

    plt.figure(figsize=(12,8))
    plt.scatter(X.lookback, y, color='r', alpha=0.9) 

    
    plt.plot(X_prime[:, 1], y_hat, 'r', alpha=0.7, label='PLS') 
    plt.plot(X_prime[:, 1], y_hat2, 'darkblue', alpha=0.9, label='PCR') 
    plt.scatter(X2.lookback, y2, color='b', alpha=0.7) 
    
    plt.xticks(ticks=X.lookback, labels=X.lookback,fontsize=16)
    # Plot the raw data 
    plt.legend()
    plt.xlabel("Lookback") 
    plt.ylabel("MSE") 
    
    plt.title('{} lookback vs MSE'.format(file2.split('-')[1][:-4]))
    # plt.savefig('/home/yicheng.wang/UCL-Dissertation/Plots/ML-model/Lookback-PLS-PCR/{}.png'.format(file2.split('-')[1][:-4]))
    plt.show()
```
